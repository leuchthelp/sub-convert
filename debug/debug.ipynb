```python
import pandas as pd

path = "debug/f0bc4f-Jujutsu Kaisen S01E14.mkv-5/f0bc4f.json"


path = path.replace("debug/", "")
df = pd.read_json(path)
```

```python
import plotly.express as px

fig = px.timeline(
    data_frame=df,
    x_start="start",
    x_end="end",
    y="placement",
    hover_name="text",
    hover_data="text",
    color="placement",
    color_discrete_map={
        "Top": "#AB029E",
        "Bottom": "#FF4430",
    },
    category_orders={"placement": ["Top", "Bottom"]},
)

fig.show()
```
